```python
# the pairplot to plot all the cloums agasint eachother 
df_plot = df[['amount','timestamp','oldbalanceOrig','newbalanceOrig','oldbalanceDest','newbalanceDest']]
# making the figure
sns.pairplot(data=df_plot,
             x_vars=['oldbalanceOrig','newbalanceOrig'],
             y_vars=['oldbalanceDest','newbalanceDest'])
sns.axes_style('whitegrid')
plt.show()
```

```python
# for plotting the entire tractions over the timestamp
# filtering the require data
df_plot = df[['amount','timestamp','oldbalanceOrig','newbalanceOrig','oldbalanceDest','newbalanceDest']]
# building the plot
sns.jointplot(data= df_plot, y=df_plot.amount,x=df_plot.timestamp,kind='scatter')
sns.axes_style('whitegrid')
plt.show()
```

```python
# for plotting the timestamps with increased transscations
data=df.loc[(df['timestamp'] > 250) & (df['timestamp'] < 350)]
sns.jointplot(data=data,x=data['timestamp'],y=data['amount'],kind='scatter')
sns.axes_style('whitegrid')
plt.show()
```
